## Data Manipulation

```python
mask = df['win_by_runs']>50
df[mask].shape
```

```python
#Create a function to filter the DataFrame for matches won by
#more than a specific number of runs
def victory(runs):
    mask = df['win_by_runs']>runs
    return df[mask]   
```

```python
# Use the 'victory' function to get number of matches won by more than 100 runs
victory(100).shape[0]
```

```python
# Create a mask for matches held in Indore after 1st Jan 2015
mask1 = df['city'] == 'Indore'
mask2 = df['date'] > '2015-01-01'

#Apply masks to df
df[mask1 & mask2]

```
