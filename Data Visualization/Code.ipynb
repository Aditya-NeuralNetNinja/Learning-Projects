## Data Visualization

```python
#Countplot of transmission types
sns.countplot(x='transmission', data=df)
plt.title('Transmission types')
plt.show()
#Dataset is highly imbalanced with respect to the manual transmission type
```

Dataset is highly imbalanced with respect to the manual transmission type

```python
#Pie chart of transmission types
fig = plt.figure(figsize=(7,7))
colors = ("gold", "red", "yellowgreen")
wp = {'linewidth':2, 'edgecolor':"black"}
tags = df['transmission'].value_counts()
explode= (0.1, 0.1, 0.1)
tags.plot(kind='pie', autopct='%1.1f%%', shadow=True, colors = colors, startangle=90, 
          wedgeprops = wp, explode=explode, label='')
plt.title('Distribution of the different transmission types')
plt.show()
```

```python
#Countplot of Fuel types
sns.countplot(y='fuelType', data=df)
plt.title('Fuel type')
plt.show()
#Dataset is highly imbalanced with respect to petrol fuelled cars
```

```python
print(df['fuelType'].value_counts())
```

```python
# Creating dataframes
fueltype = df['fuelType']
transmission = df['transmission']
price = df['price']
```

```python
# Barplots of fuel types & transmission
fig, axes = plt.subplots(1,2, figsize=(15,5), sharey=True)
fig.suptitle('Visualizing categorical data columns')
sns.barplot(x=fueltype, y=price, ax = axes[0])
sns.barplot(x=transmission, y=price, ax=axes[1])
```

```python
#Enhanced boxplots
fig, axes = plt.subplots(1,2, figsize=(15,5), sharey=True)
fig.suptitle('Visualizing categorical data columns')
sns.boxenplot(x=fueltype, y=price, ax = axes[0])
sns.boxenplot(x=transmission, y=price, ax= axes[1])
```

```python
df.corr()['price'].sort_values()
```

## Inferences:
From the correlation coefficients, we can infer the following:

1. **Mileage**: There is a moderately strong negative correlation between the mileage of a car and its price (-0.53). This suggests that, as the mileage on a car increases, its price tends to decrease, which aligns with the general expectation that used cars with higher mileage are less valuable.

2. **MPG (Miles Per Gallon)**: The correlation between MPG and price is negative but weaker (-0.35). Higher MPG means better fuel efficiency, but it seems that more fuel-efficient cars are associated with lower prices in this dataset. This might be because many high-efficiency cars are smaller models or economy vehicles, which tend to be less expensive.

3. **Fuel Type**: The correlation between fuel type and price is positive but weak (0.21). This might suggest that cars with certain fuel types are slightly more expensive than others, but the relationship is not strong. The exact interpretation would depend on how fuel type is coded in your data.

4. **Transmission**: There's a weak positive correlation between transmission and price (0.23). This suggests that certain transmission types are associated with higher prices. The exact interpretation depends on how transmission is coded in your dataset (e.g., Manual = 0, Automatic = 1).

5. **Tax**: There's a moderate positive correlation between tax and price (0.41). This suggests that more expensive cars are also associated with higher taxes.

6. **Engine Size**: The engine size has a moderate positive correlation with the price (0.41). This implies that cars with larger engines tend to be more expensive, which could be because larger engines are often found in higher-performance or luxury vehicles.

7. **Year**: There's a fairly strong positive correlation between the year of manufacture and the price (0.64). This means that newer cars are typically more expensive than older cars, as we'd generally expect in the used car market.

NOTE: Correlation does not imply causation. While these variables show a correlation with the price, it doesn't necessarily mean they directly cause the price to change. Other factors not included in this dataset may also play a significant role. Further, these correlations are based on the assumption that your data is linearly related. If there are non-linear relationships, they may not be accurately represented by these correlation coefficients.
